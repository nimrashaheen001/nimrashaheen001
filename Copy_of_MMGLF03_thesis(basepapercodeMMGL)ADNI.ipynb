```python
import os
import pandas as pd

# Path to training DICOM folders
dicom_root = "/content/MMGLF/MMGLF-zipped/dataset/train"

# Path to tabular metadata CSV
csv_path = "/content/MMGLF/MMGLF-zipped/dataset/tabular.csv"

# Output .txt path
output_txt = "/content/MMGLF/MMGLF-zipped/train-oasis.txt"

# Load tabular data
df = pd.read_csv(csv_path)

# Create train-oasis.txt
with open(output_txt, 'w') as f:
    for subject in os.listdir(dicom_root):
        subject_path = os.path.join(dicom_root, subject)
        if not os.path.isdir(subject_path):
            continue
        row = df[df["PTID"] == subject]
        if not row.empty:
            cdr = float(row.iloc[0]["CDR"])
            label = 1 if cdr >= 0.5 else 0
            f.write(f"{subject_path} {label}\n")

print("✅ Created:", output_txt)
```

```python
# For val-oasis.txt
val_dicom_root = "/content/MMGLF/MMGLF-zipped/dataset/val"
val_output_txt = "/content/MMGLF/MMGLF-zipped/val-oasis.txt"

with open(val_output_txt, 'w') as f:
    for subject in os.listdir(val_dicom_root):
        subject_path = os.path.join(val_dicom_root, subject)
        if not os.path.isdir(subject_path):
            continue
        row = df[df["PTID"] == subject]
        if not row.empty:
            cdr = float(row.iloc[0]["CDR"])
            label = 1 if cdr >= 0.5 else 0
            f.write(f"{subject_path} {label}\n")

print("✅ Created:", val_output_txt)
```

```python
import os
import pandas as pd
from glob import glob

# Define paths
base_path = "/content/MMGLF/MMGLF-zipped/dataset"
csv_path = os.path.join(base_path, "tabular.csv")
train_dir = os.path.join(base_path, "train")
val_dir = os.path.join(base_path, "val")
train_txt = "/content/MMGLF/MMGLF-zipped/train-oasis.txt"
val_txt = "/content/MMGLF/MMGLF-zipped/val-oasis.txt"

# Load CSV
df = pd.read_csv(csv_path)

# Extract filename from image_path
df['image_filename'] = df['ImageID'].apply(lambda x: os.path.splitext(os.path.basename(str(x).strip()))[0])

# Map classes to numeric labels
mapping = {
    'CN': 0,
    'MCI': 1,
    'AD': 2
}
df['label'] = df['Class'].map(mapping)
df = df.dropna(subset=['label'])  # remove unknowns
df['label'] = df['label'].astype(int)

# Helper function to create txt file
def create_txt(data_dir, output_txt, df):
    with open(output_txt, 'w') as f:
        image_files = glob(os.path.join(data_dir, "*.jpg"))
        for file_path in image_files:
            filename = os.path.splitext(os.path.basename(file_path))[0]
            if filename in df['image_filename'].values:
                label = int(df[df['image_filename'] == filename].iloc[0]['label'])
                f.write(f"{file_path} {label}\n")

# Create txt files
create_txt(train_dir, train_txt, df)
create_txt(val_dir, val_txt, df)

print("✅ Created:")
print("   -", train_txt)
print("   -", val_txt)
```
